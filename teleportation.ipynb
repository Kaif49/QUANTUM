```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
circuit=QuantumCircuit(3,3)
circuit.x(0)
circuit.h(1)
circuit.cx(1,2)
circuit.cx(0,1)
circuit.h(0)
circuit.measure([0,1],[0,1])
circuit.cx(1,2)
circuit.cz(0,2)
circuit.measure([2],[2])
circuit.draw(output='mpl')
```

```python
sim=Aer.get_backend('qasm_simulator')
result=execute(circuit,backend=sim,shots=1024).result()
plot_histogram(result.get_counts(circuit))
```

```python
#We can see the from the histogram the last digit in histogram is always one, which is copies from q0 to q2 
#in histogram  100=q2,q1,q0
```
