```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
%matplotlib inline
```

```python
##We have four numbers as 00,01,10,11 we will find 11
#We will create our own gate now called mygate=CZ
mygate=QuantumCircuit(2,name='mygate')
mygate.cz(0,1)
mygate.to_gate() #coverts into gate
mygate.draw(output='mpl')
```

```python
circuit=QuantumCircuit(2,2)
circuit.h([0,1])
circuit.append(mygate,[0,1])
circuit.draw(output='mpl')
```

```python
##Creating other gate
refcircuit=QuantumCircuit(2,name='refcircuit')
refcircuit.h([0,1])
refcircuit.z([0,1])
refcircuit.cz(0,1)
refcircuit.h([0,1])
refcircuit.to_gate()
refcircuit.draw(output='mpl')
```

```python
circuit.append(refcircuit,[0,1])
circuit.measure([0,1],[0,1])
circuit.draw(output='mpl')
```

```python
backend=Aer.get_backend('qasm_simulator')
result=execute(circuit,backend=backend,shots=1).result()
counts=result.get_counts(circuit)
plot_histogram([counts])
```
