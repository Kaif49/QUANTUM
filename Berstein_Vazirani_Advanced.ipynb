```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
%matplotlib inline
```

```python
number=input('enter number in binary')
```

```python
circuit=QuantumCircuit(len(number)+1,len(number))
```

```python
circuit.h(range(len(number)))
circuit.x(len(number))
circuit.h(len(number))
circuit.barrier()
circuit.draw(output='mpl')
```

```python
for index, one in enumerate(number[::-1]):
    print(f"index {index} is {one}")
    if one == "1":
        circuit.cx(index,len(number))

circuit.barrier()
circuit.h(range(len(number)))
circuit.barrier()
circuit.measure(range(len(number)),range(len(number)))
circuit.draw(output='mpl')
          
```

```python
simulator=Aer.get_backend('qasm_simulator')
res=execute(circuit,backend=simulator,shots=1).result()
counts=res.get_counts()
print(counts)
```

```python
plot_histogram([counts])
```
